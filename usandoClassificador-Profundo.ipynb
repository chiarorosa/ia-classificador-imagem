```python
import os
import tensorflow as tf
from tensorflow.keras.preprocessing import image
import numpy as np
```

```python
# Carregue o modelo treinado
model = tf.keras.models.load_model('./models/cfIsNoel-RCastilhos.keras')
```

```python
# DiretÃ³rio de imagens de teste
test_directory = './test'  # Substitua 'data/test' pelo caminho para a pasta de teste
```

```python
# Obtenha a lista de arquivos na pasta de teste e ordene-a alfabeticamente
image_files = sorted([
    filename for filename in os.listdir(test_directory) 
    if filename.endswith((".jpg", ".jpeg"))
])

erro_s = []
erro_ns = []

t = len(image_files)

classe0_hit = 0
classe0_miss = 0

classe1_hit = 0
classe1_miss = 0
```

```python
# Percorra todas as imagens na pasta de teste
for filename in image_files:
    # print(filename)
    
    # Caminho completo para a imagem de teste
    image_path = os.path.join(test_directory, filename)

    # PrÃ©-processe a imagem de teste
    img = image.load_img(image_path, target_size=(150, 150))
    img = image.img_to_array(img)
    img = np.expand_dims(img, axis=0)
    img = img / 255.0  # Normalize a imagem (igual ao prÃ©-processamento durante o treinamento)

    # FaÃ§a a previsÃ£o usando o modelo carregado
    prediction = model.predict(img)

    if prediction < 0.5:
        # print("Classe 0, nÃ£o Ã© Papai Noel", end='\t' )  # Defina a classe correspondente
        if 'not-a' in filename:
            # print('ACERTOU')
            classe0_hit += 1
        else:
            # print('ERROU')
            classe0_miss +=1
            erro_ns.append(filename)
    else:
        # print("Classe 1, Ã© Papai Noel ... CadÃª meu presente !?", end='\t')  # Defina a classe correspondente
        if 'not-a' not in filename:
            # print('ACERTOU')
            classe1_hit += 1
        else:
            # print('ERROU')
            classe1_miss += 1
            erro_s.append(filename)

    # Classifique a previsÃ£o
    print(f'%.2f\n' % (prediction))
```

```python
    
print(f'Tamanho da Amostra {t}')
print()
print(f'Quantidade de Acertos S  {classe1_hit} ({(classe1_hit/t)*100:2.2f}%)', end='\t')
print(f'Quantidade de Erros   S  {classe1_miss} ({(classe1_miss/t)*100:2.2f}%)')
print('Imagens que eu disse ser Papai Noel mas errei: ', erro_s)
print()
print(f'Quantidade de Acertos NS {classe0_hit} ({(classe0_hit/t)*100:2.2f}%)', end='\t')
print(f'Quantidade de Erros   NS {classe0_miss} ({(classe0_miss/t)*100:2.2f}%)')
print('Imagens que eu disse nÃ£o ser Papai Noel mas errei:', erro_ns)
```
